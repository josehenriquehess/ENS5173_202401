## 1.1 Numpy arrays

```python
import numpy as np # Importa a biblioteca numpy e a chama de np

# Criando um vetor de dados aleatórios
data = np.random.rand(1000,2) # Cria um vetor de dados chamado data, com valores aleatórios, de 1000 linhas e 2 colunas

# Verificando o tamanho do vetor
print(data.shape) # Imprime as dimensões do vetor data
print(np.size(data)) # Imprime a quantidade de células no vetor data

# Print de uma amostra dos dados
print(data[0:10]) # Imprime as 10 primeiras linhas do vetor data
print(data[0:10].shape) # Imprime as dimensões das 10 primeiras linhas do vetor data
```

```python
# Usando Monte-Carlo
dataMax = 1000 # Cria a variável dataMax com valor 1000
dataMin = 10 # Cria a variável dataMin com valor 10
data = (dataMax - dataMin)*data + dataMin # Altera o vetor data, multiplicando seus valores pela diferença de dataMax com dataMin e somando com dataMin
print(data[0:10]) # Imprime as 10 primeiras linhas do vetor data atualizado
```

```python
import matplotlib.pyplot as plt # Importa uma determinada coleção de funções do matplotlib e a chama de plt

fig,ax = plt.subplots(3) # Cria 3 subplots com fig e um conjunto de eixos ax para serem utilizados com o fim de gerar gráfico
ax[0].plot(data) # Plota no primeiro subplot os dados da primeira e da segunda coluna para o eixo y de acordo com o índice no eixo x, ligando-os através de pontos
ax[1].plot(data[:,0]) # Plota no segundo subplot os dados da primeira coluna para o eixo y de acordo com o índice no eixo x, ligando-os através de pontos
ax[2].plot(data[:,1]) # Plota no terceiro subplot os dados da primeira coluna para o eixo y de acordo com o índice no eixo x, ligando-os através de pontos
print(type(data)) # Imprime a classe do vetor data
```

```python
# Criando uma função
def noiseCreation(path,year): # Cria a função noiseCreation que tem como entrada a variável path e year
    """Esta função cria um ruido com sazonalidade.
    Inputs:
        path = asdasdas
        year = asdasdasdas

    Criado por José
    """

    # Criando um dataframe
    df = pd.DataFrame() # Através do pandas, cria um dataframe chamado df

    # Criando vertor temporal
    dtime = np.arange(np.datetime64(str(year)+"-01-01"), np.datetime64(str(year+1)+"-01-01"), np.timedelta64(1, "h"))
    # Na linha anterior, foi criado uma série de dados que vão do primeiro dia de ano até o primeiro dia de ano+1, que variarão de 1 em 1 hora

    # Montando um dataframe
    df['datetime'] = dtime # Adiciona a série de dados dtime no dataframe df com o nome da coluna datetime

    # Datetime como index do dataframe
    df = df.set_index(df['datetime']) # Altera os valores de índice do dataframe df, substituindo-os pelos valores da coluna datetime

    # Criando colunas com ano, mês e dia
    df['year'] = df.index.year # Cria a coluna year no df com base no valor de ano do índice
    df['month'] = df.index.month # Cria a coluna month no df com base no valor de mês do índice
    df['day'] = df.index.day # Cria a coluna day no df com base no valor de dia do índice

    # Crio uma coluna random
    df['random'] = np.random.rand(df.shape[0]) # Cria uma coluna no dataframe df chamada random com valores de 0 a 1, e com o número de linhas da primeira coluna do dataframe df

    # Adicionar sazonalidade
    tt = np.arange(0,df.shape[0],1) # Cria um array que vai de 0 até o tamanho do dataframe menos 1, somando de 1 em 1
    print(tt.shape) # Imprime o tamanho do array tt
    df['sazonal'] = (np.cos(tt/(360*4))) # Cria uma coluna em df com nome sazonal que é o cosseno do valor de cada linha de tt sobre o valor 360 multiplicado com 4
    df['timeSeries'] = 10*df['sazonal']+df['random'] # Cria uma coluna em df com nome timeSeries que soma 10 vezes a coluna sazonal com a coluna random
    #df['sazonal'].plot()
    df['timeSeries'].plot(color='red') # Plota a coluna timeSeries com cor vermelha variando de janeiro até dezembro de 2022
    df.to_csv(path+'noise_'+str(year)+'.csv') # Salva o dataframe df como csv com nome noise_ e em segui o ano em um determinado diretório

    return df # função retorna o dataframe df criado
```

```python
anos = [2001,2030,2032,2053,2060] # Cria uma lista com valores de anos que são 2001, 2030, 2032, 2053 e 2060

for ano in anos: # Utiliza como estrututra de repetição o for para utilizar os valores da lista anos
    print(ano) # Imprime o valor da lista que está sendo usado no momento
    if ano >2030: # Se o valor ano for maior que 2030, fará o próximo processo
        df = noiseCreation(path,ano) # Cria um dataframe df com a função noiseCreation, com dados de entrada path e ano    
```

```python
import os # Importa a biblioteca os 
files = os.listdir(path) # cria uma lista files com o nome dos arquivos no diretório path
print(files) # Imprime a lista de arquivos files

allData = [] # Cria uma lista allData com nada dentro
print(type(allData))
for file in files: # Estrutura de repetição for, que irá de arquivo em arquivo da lista files
    if file.endswith('.csv'): # Lógica, se o arquivo termina com .csv, acontecerá as próximas linhas
        print(file) # Imprime o nome do arquivo
        data = pd.read_csv(path+'/'+file) # Define a variável data como os dados do arquivo .csv
        allData.append(data) # Adiciona a allData os dados de data
```

```python
allDataDf = pd.concat(allData) # Através do pandas, cria um dataframe concatenando todos os dados de allData, ou seja, ao invés de ser uma lista com vários dataframes, agora será um data frame só, juntando todos de acordo com as colunas
```
