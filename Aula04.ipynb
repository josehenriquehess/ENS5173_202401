```python
# Criando uma lista em python
list = [0, 'Vasco', 2.5]
print(list)

# Nesta parte fiz a adição entre 1 string e 1 lista transformada em string, sendo que list[1] é o segundo índice do vetor list
print('Toda a lista: '+str(list))

# Nesta parte fiz a adição entre duas strings, sendo que list[1] é o segundo índice do vetor list
print('Variável:' + list[1])

# Adição do primeiro índice da variável (list[0] e [2]) - Lembrar que a indexação do python inicia em 0 
print(list[0] + list[2])
```

```python
# Criando uma matriz numpy
import numpy as np

# Criando uma matriz de zeros utilizando np.zeros
a = np.zeros((5,5))
print(a)
print('Indexação numpy')
print(a[0,0])
```

```python
b=np.random.rand(5,5)
print(b)
print('Dimensão da matriz')
print(b.shape)
```

```python
# Criando matriz de valores aleatórios entre uma faixa estabelecida
vmin = 1 # Valor mínimo
vmax = 100 # Valor máximo

# Matriz 
c = vmin+(vmax-vmin)*b
print(c)
```

```python
# Plotar matriz
import matplotlib.pyplot as plt

plt.plot(c)
```

```python
# Plotando a matriz transposta
plt.plot(c.transpose())
```

```python
# Plotando apenas a primeira linha
x = np.array(range(0,np.size(c[0,:].transpose())))
plt.plot(c[0,:].transpose())
plt.xticks(x)
```

```python
# Criando um gráfico de barras e exportando

fig, ax = plt.subplots(len(c[:,0]),1)
color = ['r','orange','y','g','b']

ax[0].bar(x, c[:,0], color = color)
ax[1].bar(x, c[:,1], color = color)
ax[2].bar(x, c[:,2], color = color)
ax[3].bar(x, c[:,3], color = color)
ax[4].bar(x, c[:,4], color = color)

# Exportar figura
import os
rootFolder = os.getcwd()

fig.savefig(rootFolder+r'\figures\Figura01.png')
```
